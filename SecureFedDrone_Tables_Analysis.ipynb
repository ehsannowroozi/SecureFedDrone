# SecureFedDrone Tables Analysis NotebookLoad CSVs, visualize, and export LaTeX tables.

```
import pandas as pdimport matplotlib.pyplot as plt
```

## Load CSV Files

```
attack_df = pd.read_csv("final_attack_table.csv")stealth_df = pd.read_csv("final_stealth_table.csv")robust_df = pd.read_csv("final_robust_table.csv")attack_df, stealth_df, robust_df
```

## Plot Attack Accuracy vs Beta

```
plt.figure()plt.plot(attack_df["beta"], attack_df["accuracy"], marker='o')plt.xlabel("Beta")plt.ylabel("Accuracy (%)")plt.title("Attack Strength vs Accuracy")plt.grid()plt.show()
```

## Plot Accuracy Drop

```
plt.figure()plt.bar(attack_df["beta"].astype(str), attack_df["accuracy_drop"])plt.xlabel("Beta")plt.ylabel("Accuracy Drop (pp)")plt.title("Accuracy Drop vs Beta")plt.show()
```

## Plot Stealth Relative Differences

```
plt.figure()plt.barh(stealth_df["metric"], stealth_df["relative_diff_%"])plt.xlabel("Relative Difference (%)")plt.title("Stealth Analysis")plt.show()
```

## Plot Watermark Robustness

```
plt.figure()plt.plot(robust_df["scenario"], robust_df["pass_rate"], marker='o')plt.xticks(rotation=45)plt.ylabel("Pass Rate (%)")plt.title("Watermark Robustness")plt.grid()plt.show()
```

## Export LaTeX Tables

```
attack_df.to_latex("attack_table.tex", index=False)stealth_df.to_latex("stealth_table.tex", index=False)robust_df.to_latex("robust_table.tex", index=False)print("LaTeX files exported.")
```
